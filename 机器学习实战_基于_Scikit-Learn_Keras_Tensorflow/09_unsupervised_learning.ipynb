### 9.1.1 K-Means

```python
from sklearn.datasets import make_blobs

# 5个中心点
blob_centers = np.array([[0.2,2.3],
                        [-1.5,2.3],
                        [-2.8,1.8],
                        [-2.8,2.8],
                        [-2.8,1.3]])
# 标准差
blob_std = np.array([0.4,0.3,0.1,0.1,0.1])

# 为聚类产生的数据集（围绕中心点 按照标准差外扩）
X, y = make_blobs(n_samples=2000, centers=blob_centers,cluster_std=blob_std,random_state=7)

X.shape, y.shape
```

```python
# 绘制 训练点集
def plot_clusters(X, y=None):
    plt.scatter(X[:,0], X[:,1],c=y, s=1)
    plt.xlabel("$x_1$", fontsize=16)
    plt.ylabel("$x_2$", fontsize=16, rotation=0)
    
plt.figure(figsize=(12,5))
plot_clusters(X)
plt.show()
```

```python
from sklearn.cluster import KMeans

# 集群数
k = 5
# KMeans（5 个集群）
kmeans = KMeans(n_clusters=k, random_state=42)

# 训练（聚类） 返回标签（每个实例所属集群）
y_pred = kmeans.fit_predict(X)
 
y_pred
```

```python
# 绘制 数据点集
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)
    
# 绘制 中心点
def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],marker='o', s=30, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=50, linewidths=50,
                color=cross_color, zorder=11, alpha=1)

# 绘制 决策边界
def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
 
    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)
 
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)

plt.figure(figsize=(12, 5))
plot_decision_boundaries(kmeans, X)
 
plt.show()
```

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

mnist = fetch_openml("mnist_784",version=1)
mnist.target = mnist.target.astype(np.int64)
 
X_train, X_test, y_train, y_test = train_test_split(mnist["data"],mnist["target"],random_state=42)
 
for i in (X_train, X_test, y_train, y_test):
    print(i.shape)
    
filename = "my_mnist_kmeans.data"
# （3.3）Numpy memmpy + MiniBatchKMeans 基于内存 需要时才将小文件读入内存
X_mm = np.memmap(filename, dtype="float32",mode="write",shape=X_train.shape)
X_mm[:] = X_train
 
minibatch_kmeans = MiniBatchKMeans(n_clusters=10, batch_size=10, random_state=42)
minibatch_kmeans.fit(X_mm)
```

```python
np.random.seed(42)
k = 5
n_init = 10
n_iterations = 100
batch_size = 100
init_size = 500
evaluate_on_last_n_iters = 10
 
best_kmeans = None

# （3.4）数据量过大 只能每次传入小批量数据 并使用partial_fit()训练 + MiniBatchKMeans（类似于 增量 PCA）
def load_next_batch(batch_size):
    return X[np.random.choice(len(X), batch_size, replace=False)]
 
# 额外超参数选择 十次做不同的随机初始化 选择最佳随机数的执行情况
for init in range(n_init): # 10 
    minibatch_kmeans = MiniBatchKMeans(n_clusters=k, init_size=init_size)
#     print(minibatch_kmeans)
    # 额外初始化 加载小批量数据
    X_init = load_next_batch(init_size) # 500
    # 额外初始化 小批量 聚类
    minibatch_kmeans.partial_fit(X_init)
    
    minibatch_kmeans.sum_inertia_ = 0
    for iteration in range(n_iterations): # 100
        
        X_batch = load_next_batch(batch_size) # 100
        minibatch_kmeans.partial_fit(X_batch)
        
        # 最后10个小批次 iteration>=100-10=90
        if iteration >= n_iterations - evaluate_on_last_n_iters:
            # 累加惯性值
            minibatch_kmeans.sum_inertia_ += minibatch_kmeans.inertia_
    
    # 选择最佳模型
    if (best_kmeans is None or minibatch_kmeans.sum_inertia_ < best_kmeans.sum_inertia_):
        best_kmeans = minibatch_kmeans
        
best_kmeans.score(X)
```

```python
from timeit import timeit
 
times, inertias = np.empty((100,2)), np.empty((100,2))
for k in range(1,101):
    # 普通 KMeans
    kmeans_ = KMeans(n_clusters=k, random_state=42)
    # 小批量 KMeans
    minibatch_kmeans = MiniBatchKMeans(n_clusters=k, random_state=42)
    print("\r{}/{}".format(k,100), end="")
    times[k-1,0] = timeit("kmeans_.fit(X)", number=10, globals=globals())
    times[k-1,1] = timeit("minibatch_kmeans.fit(X)",number=10,globals=globals())
    inertias[k-1,0] = kmeans_.inertia_
    inertias[k-1,1] = minibatch_kmeans.inertia_

plt.figure(figsize=(12,5))
# 左图
plt.subplot(121)
# 惯性值
plt.plot(range(1,101), inertias[:,0],"r--",label="K-Means")
plt.plot(range(1,101), inertias[:,1],"b.-",label="Mini-batch K-Means")
plt.xlabel("$k$", fontsize=16)
plt.title("Inertia", fontsize=16)
plt.legend(fontsize=14)
plt.axis([1,100,0,100])

# 右图
plt.subplot(122)
# 耗时
plt.plot(range(1,101), times[:,0],"r--",label="K-Means")
plt.plot(range(1,101), times[:,1],"b.-",label="Mini-batch K-Means")
plt.xlabel("$k$",fontsize=16)
plt.title("Training time(seconds)", fontsize=16)
plt.axis([1,100,0,6])
plt.legend(fontsize=14)
 
plt.tight_layout()
plt.show()
```

```python
# （4.1）选择 肘部 为 簇的个数

# 选择簇的个数 1-9
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(X) for k in range(1, 10)]

# 惯性值不能作为性能评估的标准
inertias = [model.inertia_ for model in kmeans_per_k]

plt.figure(figsize=(8, 5))
# 绘制 簇的个数-惯性值 曲线
plt.plot(range(1, 10), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.annotate('Elbow', xy=(4, inertias[3]), xytext=(0.55, 0.55), textcoords='figure fraction',
             fontsize=16, arrowprops=dict(facecolor='black', shrink=0.1))
plt.axis([1, 8.5, 0, 1300])

plt.show()
```

```python
from sklearn.metrics import silhouette_score

# （4.2）轮廓系数 (b-a)/max(a,b) 轮廓值越大 得分越高

# 选择簇的个数 2-9 对应的聚类模型 model 
# 每个模型 model 计算所有实例 X 及其 每个实例的标签所属集群 labels_ 的轮廓值
silhouette_scores = [silhouette_score(X, model.labels_) for model in kmeans_per_k[1:]]
 
plt.figure(figsize=(8, 5))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.axis([1.8, 8.5, 0.55, 0.7])
 
plt.show()
```

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter
 
plt.figure(figsize=(12, 10))

# 共四种集群数
for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)
    # 每个实例的标签所属集群
    y_pred = kmeans_per_k[k - 1].labels_
    # 每个实例的轮廓值
    silhouette_coefficients = silhouette_samples(X, y_pred)
 
    padding = len(X) // 30
    pos = padding
    ticks = []
    
    # 每种集群数 遍历所有集群
    for i in range(k):
        # 属于该集群的 每个实例的轮廓值
        coeffs = silhouette_coefficients[y_pred == i]
        # 排序 
        coeffs.sort()
 
        # 绘制刀形
        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding
 
    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    
    if k in (3, 5):
        plt.ylabel("Cluster")
    
    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)
 
    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)

plt.tight_layout()
plt.show()
```

### 9.1.4 使用聚类进行预处理

```python
# （6.2）有监督学习的数据预处理
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
 
X_digits, y_digits = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits, random_state=42)
 
for i in (X_train, X_test, y_train, y_test):
    print(i.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# （6.2.1）仅逻辑回归
log_reg = LogisticRegression(multi_class="ovr", solver="lbfgs", max_iter=5000, random_state=42)
log_reg.fit(X_train, y_train)
log_reg.score(X_test, y_test)
```

```python
from sklearn.pipeline import Pipeline

# （6.2.2）先聚类预处理 再逻辑回归
pipeline = Pipeline([
    # 聚类预处理（50个聚类中心）
    ("kmeans", KMeans(n_clusters=50, random_state=42)),
    # 逻辑回归
    ("log_reg", LogisticRegression(multi_class="ovr", solver="lbfgs", max_iter=5000, random_state=42)),
])
pipeline.fit(X_train, y_train)
pipeline.score(X_test, y_test)
```

```python
# 标签传播20%实例
percentile_closest = 20

X_cluster_dist = X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
# print(X_digits_dist.shape) # (1347, 50) 每个实例 与 50 个中心点的距离
# print(X_cluster_dist.shape) # (1347, )  每个实例 距离最近中心点的距离

# 对于每个聚类中心
for i in range(k):
    
    # 每个实例 是否属于 该聚类中心
    in_cluster = (kmeans.labels_ == i)
    # print("in_cluster.shape",in_cluster.shape) # (1347, ) True or False
    # print(in_cluster)
    
    # 选择 属于该聚类中心的 样本
    cluster_dist = X_cluster_dist[in_cluster]
    # print("cluster_dist.shape",cluster_dist.shape) # (属于该聚类中心的样本量，) 属于该聚类中心的实例 距离最近中心点的距离
    # print(cluster_dist)
    # print('-------------------------')
    
    # 获得 距离该聚类中心 分位20%（由小到大-由近到远）的 样本距离
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    # print(cutoff_distance)
    
    # 每个实例 距离该聚类中心 是否大于分位20%
    above_cutoff = (X_cluster_dist > cutoff_distance)
    # print(above_cutoff) # (1347, ) True or False
    
    # 剔除 离群点
    X_cluster_dist[in_cluster & above_cutoff] = -1

# 获得部分传播的样本
partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]

# 逻辑回归
log_reg = LogisticRegression(multi_class="ovr", solver="lbfgs", max_iter=5000, random_state=42)
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
log_reg.score(X_test, y_test)
```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    # 核心实例
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    # 异常值
    anomalies_mask = dbscan.labels_ == -1
    # 非核心实例
    non_core_mask = ~(core_mask | anomalies_mask)
 
    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    # 核心实例构成的集群
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    # 异常值
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    # 非核心实例
    plt.scatter(non_cores[:, 0], non_cores[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
    
plt.figure(figsize=(12, 5))
 
plt.subplot(121)
plot_dbscan(dbscan, X, size=100)
 
plt.subplot(122)
plot_dbscan(dbscan2, X, size=600, show_ylabels=False)
 
plt.tight_layout()
plt.show()
```

```python
resolution = 100

grid = np.arange(-10, 10, 1 / resolution)
xx, yy = np.meshgrid(grid, grid)
X_full = np.vstack([xx.ravel(), yy.ravel()]).T

# （8.7.1）对这些输出score值求e次幂 则可以得到给定样本位置的PDF值
pdf = np.exp(gm.score_samples(X_full))

# （8.7.2）所有高斯模型的PDF和为1
pdf_probas = pdf * (1 / resolution) ** 2
pdf_probas.sum()
```

```python
from matplotlib.colors import LogNorm
 
def plot_gaussian_mixture(clusterer, X, resolution=1000, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),np.linspace(mins[1], maxs[1], resolution))
    
    # 新增
    # 估计任何位置（X）的模型密度的对数（logPDF）（φ→X）
    Z = -clusterer.score_samples(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # 绘制 模型密度对数轮廓
    plt.contourf(xx, yy, Z, norm=LogNorm(vmin=1.0, vmax=30.0), levels=np.logspace(0, 2, 12))
    # 填充 模型密度对数轮廓
    plt.contour(xx, yy, Z, norm=LogNorm(vmin=1.0, vmax=30.0), levels=np.logspace(0, 2, 12), linewidths=1, colors='k')

    # 预测 标签所属集群
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # 绘制 预测轮廓
    plt.contour(xx, yy, Z, linewidths=2, colors='r', linestyles='dashed')
    
    # 绘制 实例样本点
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)
    # 绘制 聚类中心点 指定了权重参数φ
    plot_centroids(clusterer.means_, clusterer.weights_)
 
    plt.xlabel("$x_1$", fontsize=14)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
        
plt.figure(figsize=(8, 4))
plot_gaussian_mixture(gm, X)
 
plt.show()
```

### 9.2.2 选择聚类数

```python
from scipy.stats import norm
from matplotlib.patches import Polygon
 
# x
xx = np.linspace(-6, 4, 101)
# print(xx.shape) # (101, )
# 参数θ
ss = np.linspace(1, 2, 101)
# print(ss.shape) # (101, )
XX, SS = np.meshgrid(xx, ss)
# print(XX.shape) # (101, 101)
# print(SS.shape) # (101, 101)

# 正态分布 PDF值(L^)
ZZ = 2 * norm.pdf(XX - 1.0, 0, SS) + norm.pdf(XX + 4.0, 0, SS)
ZZ = ZZ / ZZ.sum(axis=1)[:,np.newaxis] / (xx[1] - xx[0])
# print(ZZ.shape) # (101, 101)

plt.figure(figsize=(12, 8))

# 某样本 x=85 θ=30
x_idx = 85 # 85/100 (2.5+6)/10 相当于 x=2.5
s_idx = 30 # 30/100 (1.3-1)/10 相当于 θ=1.3

# ----------------------------------------------------------------

# 左一
plt.subplot(221)
# 填充 PDF值(L^)分布 XX, SS, ZZ 
plt.contourf(XX, SS, ZZ, cmap="GnBu")
# 水平 θ=1.3
plt.plot([-6, 4], [ss[s_idx], ss[s_idx]], "k-", linewidth=2)
# 垂直 x=2.5
plt.plot([xx[x_idx], xx[x_idx]], [1, 2], "b-", linewidth=2)
plt.xlabel(r"$x$")
plt.ylabel(r"$\theta$", fontsize=14, rotation=0)
plt.title(r"Model $f(x; \theta)$", fontsize=14)

# ----------------------------------------------------------------

# 右一
plt.subplot(222)
# x=2.5 时 所有θ对应的PDF值(L^)
plt.plot(ss, ZZ[:, x_idx], "b-") # print(ZZ[:, x_idx].shape) # (101, ) 
# x=2.5 时 最大θ对应的PDF值(L^)索引
max_idx = np.argmax(ZZ[:, x_idx]) # print(max_idx) # 66
# x=2.5 时 最大θ对应的PDF值(L^)
max_val = np.max(ZZ[:, x_idx]) # print(max_val) # 0.113133068425285

# max点 
plt.plot(ss[max_idx], max_val, "r.") # print(ss[max_idx]) # 1.6600000000000001 
# max 垂直
plt.plot([ss[max_idx], ss[max_idx]], [0, max_val], "r:")
# max 水平
plt.plot([0, ss[max_idx]], [max_val, max_val], "r:")

plt.text(1.01, max_val + 0.005, r"$\hat{L}$", fontsize=14)
plt.text(ss[max_idx]+ 0.01, 0.055, r"$\hat{\theta}$", fontsize=14)
plt.text(ss[max_idx]+ 0.01, max_val - 0.012, r"$Max$", fontsize=12)
plt.axis([1, 2, 0.05, 0.15])
plt.xlabel(r"$\theta$", fontsize=14)
plt.grid(True)
plt.text(1.99, 0.135, r"$=f(x=2.5; \theta)$", fontsize=14, ha="right")
plt.title(r"Likelihood function $\mathcal{L}(\theta|x=2.5)$", fontsize=14)

# ----------------------------------------------------------------

# 左二
plt.subplot(223)

# 绘制 PDF值(L^)轮廓 XX, ZZ[s_idx] 确定θ=1.3
plt.plot(xx, ZZ[s_idx], "k-")

# 41/100 (-2+6)/10 相当于 x=-2
# 81/100 (2+6)/10 相当于 x=2
# x轴积分范围 [-2, 2] → [(xx[41], 0)] [(xx[80], 0)]
# y轴数值结果 θ=1.3时 x∈[-2, 2]的PDF值(L^) → ZZ[s_idx, 41:81]
verts = [(xx[41], 0)] + list(zip(xx[41:81], ZZ[s_idx, 41:81])) + [(xx[80], 0)]

# 函数积分图构造器 Polygon 生成阴影部分 表示积分面积
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')

# 创建的图形对象不会直接在 figure 中显示 需要添加进 axis
plt.gca().add_patch(poly)

plt.axis([-6, 4, 0, 0.25])
plt.xlabel(r"$x$", fontsize=14)
plt.grid(True)
plt.title(r"PDF $f(x; \theta=1.3)$", fontsize=14)

# ----------------------------------------------------------------

# 右二
plt.subplot(224)
# x=2.5 时 所有θ对应的logPDF值(logL^)
plt.plot(ss, np.log(ZZ[:, x_idx]), "b-")
# x=2.5 时 最大θ对应的logPDF值(logL^)索引
max_idx = np.argmax(np.log(ZZ[:, x_idx]))
# x=2.5 时 最大θ对应的logPDF值(logL^)
max_val = np.max(np.log(ZZ[:, x_idx]))

# max点 
plt.plot(ss[max_idx], max_val, "r.")
# max 垂直
plt.plot([ss[max_idx], ss[max_idx]], [-5, max_val], "r:")
# max 水平
plt.plot([0, ss[max_idx]], [max_val, max_val], "r:")

plt.axis([1, 2, -2.4, -2])
plt.xlabel(r"$\theta$", fontsize=14)
plt.text(ss[max_idx]+ 0.01, max_val - 0.05, r"$Max$", fontsize=12)
plt.text(ss[max_idx]+ 0.01, -2.39, r"$\hat{\theta}$", fontsize=14)
plt.text(1.01, max_val + 0.02, r"$\log \, \hat{L}$", fontsize=14)
plt.grid(True)
plt.title(r"$\log \, \mathcal{L}(\theta|x=2.5)$", fontsize=14)

# ----------------------------------------------------------------

plt.tight_layout()
plt.show()
```

```python
# （10.3）手动计算 BIC AIC

# n_params 参数个数
n_clusters = 3
n_dims = 2
n_params_for_weights = n_clusters - 1
n_params_for_means = n_clusters * n_dims
n_params_for_covariance = n_clusters * n_dims * (n_dims + 1) // 2
n_params = n_params_for_weights + n_params_for_means + n_params_for_covariance
# print(n_params) # 17

# log(L^) 
max_log_likelihood = gm.score(X) * len(X) 

# BIC = log(m) * p - 2 * log(L^) (m=len(X) p=n_params log(L^)=max_log_likelihood)
bic = np.log(len(X)) * n_params - 2 * max_log_likelihood

# AIC = 2 * p - 2 * log(L^)
aic = 2 * n_params - 2 * max_log_likelihood
print(bic, aic)
```

```python
# （10.4）选择最佳超参数

# （10.4.1）不同的超参数k 训练高斯混合模型

# k=1-10 训练不同的高斯混合模型
gms_per_k = [GaussianMixture(n_components=k, n_init=10, random_state=42).fit(X) for k in range(1, 11)]

# 每个模型的 BIC
bics = [model.bic(X) for model in gms_per_k]
# 每个模型的 AIC
aics = [model.aic(X) for model in gms_per_k]
 
plt.figure(figsize=(6, 4))

# 绘制 k-BIC
plt.plot(range(1, 11), bics, "bo-", label="BIC")
# 绘制 k-AIC
plt.plot(range(1, 11), aics, "go--", label="AIC")

plt.xlabel("$k$", fontsize=14)
plt.ylabel("Information Criterion", fontsize=14)
plt.axis([1, 9.5, np.min(aics) - 50, np.max(aics) + 50])
plt.annotate('Minimum',xy=(3, bics[2]),xytext=(0.35, 0.6),textcoords='figure fraction',fontsize=14,
             arrowprops=dict(facecolor='black', shrink=0.1))
plt.legend()
 
plt.show()
```

```python
# （10.4.2）不同的超参数 covariance_type

min_bic = np.infty

# k=1-10 
for k in range(1, 11):
    
    # covariance_type = "full", "tied", "spherical", "diag" 训练不同的高斯混合模型
    for covariance_type in ("full", "tied", "spherical", "diag"):
        bic = GaussianMixture(n_components=k, n_init=10,covariance_type=covariance_type,random_state=42).fit(X).bic(X)
    
        # 存储最小BIC 对应的 最优参数（k、covariance_type）
        if bic < min_bic:
            min_bic = bic
            best_k = k
            best_covariance_type = covariance_type
            
print(best_k, best_covariance_type)
```
