## 4.1 线性回归

```python
import numpy as np
import matplotlib.pyplot as plt

# 随机生成100个点
X = 2 * np.random.rand(100, 1)
y = 4+3 * X + np.random.randn(100,1)

plt.plot(X, y, "b.")
plt.show()
```

```python
# 加上 x0
X_b = np.c_[np.ones((100,1)), X]
# （1）正规方程 求最优参数
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
# 根据最优参数 预测 y
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0,2,0,15])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# 线性回归模型
lin_reg = LinearRegression()
# 训练
lin_reg.fit(X, y)
# （2.1）线性回归模型 求最优参数
lin_reg.intercept_, lin_reg.coef_
```

```python
# 根据最优参数 预测 y
lin_reg.predict(X_new)
```

```python
# （2.2）最小二乘 np.linalg.lstsq 求最优参数
theta_best_svd, residuals, rank, s = np.linalg.lstsq(X_b, y, rcond=1e-6)
theta_best_svd
```

```python
# （2.3）最小二乘 X+y 求最优参数
np.linalg.pinv(X_b).dot(y)
```

## 4.2 梯度下降

### 4.2.1 批量梯度下降

```python
eta = 0.1
n_iterations = 1000
m=100

# 随机初始化参数
theta = np.random.rand(2,1)

# （3）批量梯度下降 求最优参数
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta -= eta * gradients
    
theta
```

```python
X_new_b.dot(theta)
```

```python
theta_path_bgd = []
 
def plot_gradient_descent(theta, eta, theta_path=None):
    m = len(X_b)
    # 数据 点集
    plt.plot(X, y, "b.")
    
    n_iterations = 1000
    for iteration in range(n_iterations):
        # 前十轮 线集
        if iteration < 10:
            y_predict = X_new_b.dot(theta)
            # 第一轮 红线 theta初始化时的值 | 后九轮 蓝线 theta训练九轮后的值
            style = "b-" if iteration > 0 else "r--"
            # 仅预测两个点 确定预测直线
            plt.plot(X_new, y_predict, style)
            
        # 批量梯度下降
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        # 更新参数
        theta = theta - eta * gradients
        
        if theta_path is not None:
            theta_path.append(theta)
    
    plt.xlabel("$x_1$",fontsize=18)
    plt.axis([0,2,0,15])
    plt.title(r"$\eta = {}$".format(eta), fontsize=16)
    
np.random.seed(42)
theta = np.random.randn(2,1)
 
plt.figure(figsize=(12,5))
# 学习率 = 0.02
plt.subplot(131);plot_gradient_descent(theta, eta=0.02)
plt.ylabel("$y$", rotation=0, fontsize=18)
# 学习率 = 0.1
plt.subplot(132);plot_gradient_descent(theta, eta=0.1,theta_path=theta_path_bgd)
# 学习率 = 0.5
plt.subplot(133);plot_gradient_descent(theta, eta=0.5)
 
plt.tight_layout()
plt.show()
```

### 4.2.2 随机梯度下降

```python
theta_path_sgd = []
m = len(X_b)
np.random.seed(42)
 
n_epochs = 50
t0, t1 = 5,50
 
# 学习率调度 逐步降低学习率
def learning_schedule(t):
    return t0/(t+t1)

# 随机初始化
theta = np.random.randn(2,1)

# （4.1）随机梯度下降 手动实现 求最优参数
for epoch in range(n_epochs):
    # 共m回合 每一回合选取1个样本进行随机梯度下降 并更新参数
    for i in range(m):
        # 前二十轮 线集
        if epoch == 0 and i < 20:
            y_predict = X_new_b.dot(theta)
            # 第一轮 红线 theta初始化时的值 | 后十九轮 蓝线 theta训练十九轮后的值
            style = "b-" if i > 0 else "r--"
            # 仅预测两个点 确定预测直线
            plt.plot(X_new, y_predict, style)
        
        # 随机选择一个样本
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        
        # 随机梯度下降
        gradients = 2*xi.T.dot(xi.dot(theta)-yi)
        # 更新参数
        eta = learning_schedule(epoch*m+i)
        theta = theta - eta * gradients
        
        theta_path_sgd.append(theta)

# 数据 点集
plt.plot(X,y,"b.")
plt.xlabel("$x_1$",fontsize=18)
plt.ylabel("$y$",rotation=0,fontsize=18)
plt.axis([0,2,0,15])
plt.show()

theta
```

```python
# （4.2）随机梯度下降  SGDRegressor模型 求最优参数
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1, random_state=42)
sgd_reg.fit(X,y.ravel())

sgd_reg.intercept_, sgd_reg.coef_
```

### 4.2.3 小批量梯度下降

```python
theta_path_mgd = []
n_epochs = 50
minibatch_size = 20
 
np.random.seed(42)
theta = np.random.randn(2,1)
 
t0, t1 = 200, 1000
def learning_schedule(t):
    return t0/(t+t1)
 
t = 0
# （5）小批量梯度下降 手动实现 求最优参数
for epoch in range(n_epochs):
        
    # 随机混洗
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    
    # 共m/minibatch_size回合 每个回合选取minibatch_size个样本进行小批量梯度下降 并更新参数
    for i in range(0,m,minibatch_size):
        t += 1
        
        # 随机选择minibatch_size个样本
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        
        # 随机梯度下降
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta)-yi)
        # 更新参数
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)

theta
```

```python
theta_list = [theta_path_bgd, theta_path_mgd, theta_path_sgd]
theta_type = ["Batch","Mini-batch","Stochastic"]
theta_style = ["b-o","g-+","r-s"]
  
plt.figure(figsize=(8,5))
for i in reversed(range(3)):
    plt.plot(np.array(theta_list[i])[:,0],np.array(theta_list[i])[:,1],theta_style[i],
             linewidth=2,label=theta_type[i])
    
plt.xlabel(r"$\theta_0$",fontsize=20)
plt.ylabel(r"$\theta_1$",fontsize=20, rotation=0)
plt.axis([1,5,1,5])
plt.legend(fontsize=16,loc="upper left")
plt.show()
```

## 4.4 学习曲线

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# 一阶 二阶 三百阶
for style, width, degree in (("g-", 1, 300), ("b--", 2, 2), ("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    # 流水线：加阶、标准化、线性回归
    polynomial_regression = Pipeline([("poly_features", polybig_features),
                                      ("std_scaler", std_scaler),
                                      ("lin_reg", lin_reg)])
    # 训练
    polynomial_regression.fit(X, y)
    # 预测
    y_newbig = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_newbig, style, label=str(degree), linewidth=width)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper center")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
 
# 绘制学习曲线
def plot_learning_curves(model, X, y):
    # 训练集80 验证集20
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, 
                                                      random_state=10)
    train_errors, val_errors = [], []
    # 79轮 每轮训练集大小1-m(1 2 3 ... 79)
    for m in range(1, len(X_train)):
        # 训练
        model.fit(X_train[:m], y_train[:m])
        # 预测训练集
        y_train_predict = model.predict(X_train[:m])
        # 预测验证集
        y_val_predict = model.predict(X_val)
        # 添加训练集误差
        train_errors.append(mean_squared_error(y_train[:m],y_train_predict))
        # 添加测试集误差
        val_errors.append(mean_squared_error(y_val, y_val_predict))
        
    # 训练集均方根误差
    plt.plot(np.sqrt(train_errors),"r-+",linewidth=2,label="train")
    # 验证集均方根误差
    plt.plot(np.sqrt(val_errors),"b-",linewidth=3,label="val")
    plt.legend()
    plt.xlabel("Training set size")
    plt.ylabel("RMSE")

# 线性模型（欠拟合）
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
plt.show()
```

```python
from sklearn.pipeline import Pipeline

# 多项式回归 十阶（过拟合）
polynomial_regression = Pipeline([("poly_features",PolynomialFeatures(degree=10,include_bias=False)),
                                  ("lin_reg",LinearRegression()) ])
 
plot_learning_curves(polynomial_regression, X, y)
plt.axis([0,80,0,3])
plt.show()
```

## 4.5 正则化线性模型

### 4.5.1 岭回归

```python
np.random.seed(42)

# 20个训练样本
m = 20
X = 3 * np.random.rand(m,1)
y = 1+0.5*X+np.random.randn(m,1)/1.5

# 200个测试样本
X_new = np.linspace(0,3,100).reshape(100,1)
```

```python
# https://blog.csdn.net/qq_16000815/article/details/80928728
# solver:字符串 指定求解最优化问题的算法 
#     (1).solver='auto',根据数据集自动选择算法
#     (2).solver='svd',采用奇异值分解的方法
#     (3).solver='cholesky',采用scipy.linalg.solve函数
#     (4).solver='sparse_cg',才哟个scipy.sparse.linalg.cg函数
#     (5).solver='sag',采用Stochastic Average Gradient descent算法
from sklearn.linear_model import Ridge

# solver="cholesky" 
ridge_reg = Ridge(alpha=1, solver="cholesky", random_state=42)
# 正则化后 训练模型
ridge_reg.fit(X,y)
ridge_reg.predict([[1.5]])
```

```python
# solver="sag"
ridge_reg = Ridge(alpha=1, solver="sag", random_state=42)
# 正则化后 训练模型
ridge_reg.fit(X,y)
ridge_reg.predict([[1.5]])
```

```python
# 绘制 不同模型、不同学习率的 假设函数曲线
def plot_model(model_class, polynomial, alphas, **model_kargs):
    
    # 三种不同的学习率
    for alpha, style in zip(alphas, ("b-","g--","r:")):
        # 纯线性 or +岭回归（α=0 纯线性 α>0 有正则化）
        model = model_class(alpha, **model_kargs) if alpha > 0 else LinearRegression()
        print(model)
        # +多项式
        if polynomial:
            model = Pipeline([
                ("poly_features",PolynomialFeatures(degree=10, include_bias=False)),
                ("std_scaler",StandardScaler()),
                ("regul_reg",model)
            ])
        # 训练
        model.fit(X, y)
        # 预测
        y_new_regul = model.predict(X_new)
        
        lw = 2 if alpha > 0 else 1
        # 绘制
        plt.plot(X_new, y_new_regul, style, linewidth=lw, label=r"$\alpha = {}$".format(alpha))
        
    plt.plot(X,y,"b.",linewidth=3)
    plt.legend(fontsize=18)
    plt.xlabel("$x_1$",fontsize=18)
    plt.axis([0,3,0,4])
    
plt.figure(figsize=(10,5))

# 岭回归 第一张图：仅线性回归 学习率：0 10 100
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0,10,100), random_state=42)
plt.ylabel("$y$",fontsize=18,rotation=0)

# 岭回归 第二张图：多项式回归 学习率：0 10-5 1
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0,10**-5, 1), random_state=42)
 
plt.show()
```

### 4.5.4 提前停止

```python
np.random.seed(42)

# 100个样本
m = 100
X = 6 * np.random.rand(m,1) - 3
y = 2 + X + 0.5 * X**2 + np.random.randn(m,1)
 
# 25个训练集 25个验证集
X_train, X_val, y_train, y_val = train_test_split(X[:50],y[:50],
                                                 test_size=0.5,
                                                 random_state=42)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.base import clone

# 高阶多项式回归
poly_scaler = Pipeline([
    ("poly_features",PolynomialFeatures(degree=90,include_bias=False)),
    ("std_scaler",StandardScaler())
])
# 处理训练集
X_train_poly_scaled = poly_scaler.fit_transform(X_train)
# 处理测试集
X_val_poly_scaled = poly_scaler.transform(X_val)
 
# 普通随机梯度下降（未正则化）
sgd_reg = SGDRegressor(max_iter=1,tol=-np.infty,warm_start=True,
                      penalty=None, learning_rate="constant",
                      eta0=0.0005,random_state=42)

# 初始化最小验证误差 无穷大 随后逐步减小 每次保留更小的验证误差、模型
minimum_val_error = float("inf")

best_epoch = None
best_model = None

# 增量训练模型
for epoch in range(1000):
    # 增量训练
    sgd_reg.fit(X_train_poly_scaled, y_train.ravel())
    # 预测
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    # 计算验证误差
    val_error = mean_squared_error(y_val, y_val_predict)
    # 不断更新验证误差的最小值 并存储 最佳轮数、最优模型
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)

best_epoch, best_model
```

```python
sgd_reg = SGDRegressor(max_iter=1, tol=-np.infty, warm_start=True,
                       penalty=None, learning_rate="constant", 
                       eta0=0.0005, random_state=42)
 
n_epochs = 5000
train_errors, val_errors = [], []
for epoch in range(n_epochs):
    # 训练
    sgd_reg.fit(X_train_poly_scaled, y_train.ravel())
    # 预测训练集
    y_train_predict = sgd_reg.predict(X_train_poly_scaled)
    # 预测验证集
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    # 添加训练集误差
    train_errors.append(mean_squared_error(y_train, y_train_predict))
    # 添加测试集误差
    val_errors.append(mean_squared_error(y_val, y_val_predict))

# 最佳轮数 验证误差最小时的索引
best_epoch = np.argmin(val_errors)
# 最优均方根损失值 最佳轮数对应的验证误差求平方根
best_val_rmse = np.sqrt(val_errors[best_epoch])

# 标注文字
plt.annotate('Best model',
             xy=(best_epoch, best_val_rmse), # 被注释的坐标位置
             xytext=(best_epoch, best_val_rmse + 1), # 注释文字的坐标位置
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=16,
            )

best_val_rmse -= 0.03  # just to make the graph look better

# 均方根最小值的水平线
plt.plot([0, n_epochs], [best_val_rmse, best_val_rmse], "k:", linewidth=2)

# 验证集误差
plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="Validation set")

# 训练集误差
plt.plot(np.sqrt(train_errors), "r--", linewidth=2, label="Training set")

plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("RMSE", fontsize=14)
plt.show()
```

### 4.6.3 决策边界

```python
from sklearn.linear_model import LogisticRegression

# 花瓣长度、花瓣宽度
X = iris["data"][:, (2, 3)]  
# 是否是种类二（y=0 不是种类二 y=1 是种类二）
y = (iris["target"] == 2).astype(np.int)

# 创建逻辑回归模型
log_reg = LogisticRegression(solver="lbfgs", C=10**10, random_state=42) #正则惩罚系数
# 训练
log_reg.fit(X, y)

# 测试集 x0∈(2.9,7) x1∈(0.8,2.7)
x0, x1 = np.meshgrid(
        np.linspace(2.9, 7, 500).reshape(-1, 1),
        np.linspace(0.8, 2.7, 200).reshape(-1, 1),
    )
# print(x0) # (200 行, 500 列)
# print(x1) # (200 x1, 500 x0)
X_new = np.c_[x0.ravel(), x1.ravel()]
# print(X_new.shape) # (100000, 2)
# 预测 
y_proba = log_reg.predict_proba(X_new)
 
plt.figure(figsize=(10, 4))

# y=0 （x-非鸾尾花的宽度x0 y-非鸾尾花的高度x1）
plt.plot(X[y==0, 0], X[y==0, 1], "bs") 
# print(X[y==0, 0].shape)
# print(X[y==0, 0])
# print(X[y==0, 1].shape)
# print(X[y==0, 1])
# y=1（x-鸾尾花的宽度x0 y-鸾尾花的高度x1）
plt.plot(X[y==1, 0], X[y==1, 1], "g^")
# print(X[y==1, 0].shape)
# print(X[y==1, 0])
# print(X[y==1, 1].shape)
# print(X[y==1, 1])
plt.text(3.5, 1.5, "Not Iris virginica", fontsize=14, color="b", ha="center")
plt.text(6.5, 2.3, "Iris virginica", fontsize=14, color="g", ha="center")

# 预测值概率（10000个样本）形状与x0 x1保持一致(200, 500)
zz = y_proba[:, 1].reshape(x0.shape)
# 绘制三维等高线图 轮廓线（代表 行x1 列x0 处的预测值概率）
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)
# inline=True 高度写在等高线上
plt.clabel(contour, inline=1, fontsize=12)

# 决策边界 x 两个花瓣长度 x0=2.9 and 7
left_right = np.array([2.9, 7])
# print(log_reg.coef_.shape) # (1θ0 , 2θ1) 
# print(log_reg.coef_[0][0]) # θ0
# print(log_reg.coef_[0][1]) # θ1
# 决策边界 y 两个花瓣宽度 x1 = -(θ0 * x0 + b) / θ1 （y=0）
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]
# print(boundary)
# 绘制决策边界
plt.plot(left_right, boundary, "k--", linewidth=3)

plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.axis([2.9, 7, 0.8, 2.7])
 
plt.show()
```

### 4.6.4 Softmax回归

```python
# 花瓣长度、花瓣宽度
X = iris["data"][:, (2,3)]  
# 二元→三元（多元分类）
y = iris["target"]
# 创建Softmax回归模型
softmax_reg = LogisticRegression(multi_class="multinomial",solver="lbfgs",
                                C=10,random_state=42)
# 训练
softmax_reg.fit(X,y)
```

```python
# 预测分类结果
softmax_reg.predict([[5,2]])
```

```python
# 预测每个类的概率
softmax_reg.predict_proba([[5,2]])
```
